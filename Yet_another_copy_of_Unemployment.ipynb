**Import the Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

**Loading the datasets**

```python
data = pd.read_csv('/content/Unemployment_Rate_upto_11_2020 - Unemployment_Rate_upto_11_2020.csv')
data
```

```python
#checking dataset information
data.info()
```

```python
#describe the dataset
data.describe()
```

```python
#check null and missing values
data.isnull().sum()
```

```python
# rename columns
data.columns = ['States','Date','Frequency','Estimate Unemployement Rate','Estimated Employed','Estimated Labour Participation Rate','Region','Longitude','Latitude']
```

```python
# analysing top rows of dataset
data.head()
```

**CHECKING THE CORRELATION BETWEEN THE FEATURE OF DATASET**

```python
# PLOTTING CORRELATION HEATMAP

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(9,7))
#compute the correlation matrix and plot the heatmap
plt.style.use('seaborn-v0_8-whitegrid')
sns.heatmap(data.corr(numeric_only= True))

plt.show()
```

**Estimated number of employee according to different region of India**

```python
# plotting hist plot
data.column=['State','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region','Longitude','Latitude']
plt.title('Indian Unemployement')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

**Unemployement rate according to differnt region of India**

```python
#plotting histplot
plt.figure(figsize=(12,10))
plt.title("Indian Unemployement")
sns.histplot(x="Estimate Unemployement Rate",hue="Region",data=data)
plt.show()
```

**Dashboard to analayze the unemployement rate of each India state**

```python
# plottting sunburst
from pathlib import Path
unemployement = data[['States','Region','Estimate Unemployement Rate']]
Path('/content/Unemploment_Rate_upto_11_2020 - Unemployment_Rate_upto_11_2020.csv')
figure = px.sunburst(unemployement,path=['Region','States'],
                     values = 'Estimate Unemployement Rate',
                     width=700,height=600,color_continuous_scale= 'RdY1Gn',
                     title="Unemployement Rate in India")
figure.write_html("unemployment_rate_sunburst.html")
figure.show()
```
